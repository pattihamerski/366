In this assignment, we will begin creating plots using the Python library Matplotlib. The import statement has been included for you. Run the code cell below to gain access to NumPy functions through `np` and Matplotlib functions (specifically, Matplotlib's pyplot interface) through `plt`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Making a plot in Matplotlib comes with a basic checklist:
1. An array of $x$-values
2. An equally sized array of $y$-values (if plotting on 2 axes)
3. A plotting command, to create the plot
4. A `show` command, to display the plot in your interface

For instance, if we want to plot $\sin(x)$ in the range $0 \le x \le 2\pi$, we would first create an array of values for $x$:

```python
x = np.linspace(0, 2*np.pi, 100)
```

Note that computers always work with a finite number of values, so you cannot really plot $\sin(x)$ *continuously* in the range $0 \le x \le 2\pi$. What we can do is "discretize" $x$ within the desired range. The NumPy command above, for example, creates 100 values for $x$ in the range $\left[0,2\pi\right]$. 

Once you have your independent variable, you can create corresponding y-values. Because we need an equal number of x and y values, we can use NumPy operations, which preserve the size of an array.

```python
y = np.sin(x)
```

Now that we have equally sized x and y arrays, we can plot them. If you want to emphasize the discrete nature of the arrays, `scatter` is a good command.

```python
# plotting command "scatter" creates a scatterplot
plt.scatter(x, y)
# plt.show() shows the plot in your interface
plt.show()
```

Alternatively, you can use `plot` to instruct the computer to draw a line between each subsequent point, creating the illusion of a continuous curve.

```python
# plotting command "plot" creates a connected curve
plt.plot(x, y)
# plt.show() shows the plot in your interface
plt.show()
```

Lastly, we can add labels to our plot to make the information and purpose more clear to whoever is viewing it. The computer will create a plot regardless, but you should **always** add labels for visual clarity, unless they would add unnecessary clutter.

```python
plt.plot(x, y)

# xlabel and ylabel create axis labels
plt.xlabel("x")
plt.ylabel("sin(x)")

# title creates a title over the top of the plot
plt.title("Example of sine wave")

# always place the "show" command at the very end
plt.show()
```
