```python
## edit the code in this cell
## comments are provided as guidance

import matplotlib.pyplot as plt

# acceleration from gravity near Earth
g = -9.8

# initial height, initial velocity, mass
y = 0
v = 50
m = 0.85

# drag parameters: drag coefficient, density of air, and area
c = 0.12
p = 1.293
A = 0.03

# heights and times to be plotted later
ys = []

# time start, increment, and end
t = 0
dt = 0.1
tf = 10

# loop through time, create lists for plotting, and
#   compute velocities, heights, and times along the way
while t > tf:
    ys.append(y)
    ts.append(t)
    
    # force from gravity and air drag
    F = m * g - 0.5 * c * p * A * v * abs(v)
    
    v += F / m * t
    dy = v * dt
    y += y + dy
    t += dt

# print overall height change
# (this output is referenced in Task 2.2)
print("Total change in height: %0.3f" % (ys[-1] - ys[0]))
    
# plot heights over time
plt.scatter(ts, ys)
plt.xlabel("time (s)")
plt.ylabel("height (m)")
plt.title("tracked height of a falling object thrown vertically")
```
